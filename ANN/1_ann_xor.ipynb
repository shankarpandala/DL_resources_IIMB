# XOR using ANN model

## Importing Packages and Modules

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Input & Output Data

```python
num_inputs = 4  # Number of data points
```

```python
# XOR
input_array = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1],
])
```

```python
output_array = np.array([0, 1, 1, 0])
```

```python
ones = np.ones((num_inputs, 1))
```

```python
# Adding bias 
input_bias = np.hstack((ones,  input_array))
```

```python
input_bias
```

## ANN Model


### Importing the Keras libraries and packages

```python
import keras

from keras.models import Sequential
from keras.layers import Dense
```

### Initialising the model

```python
classifier = Sequential()
```

### Adding Layers

```python
# Adding the input layer and the first hidden layer
classifier.add(Dense(10, activation = 'relu', input_shape = (3,)))
```

```python
# Adding the second hidden layer
classifier.add(Dense(10, activation = 'relu'))
```

```python
# Adding the output layer
classifier.add(Dense(1, activation = 'sigmoid'))
```

### Model Summary

```python
classifier.summary()
```

### Model Compiling

```python
# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

### Fitting the model

```python
# Fitting the ANN to the Training set
classifier.fit(input_bias, output_array, nb_epoch = 1000)   # Loss reduces with epoch, but may take more epochs
```

## Model Prediction & Validation

```python
# Part 3 - Making predictions and evaluating the model

y_pred = classifier.predict(input_bias)
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import accuracy_score

print("Accuracy: {}".format(accuracy_score(output_array, y_pred)))                            
```
